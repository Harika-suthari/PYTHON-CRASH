Append

-- Adding elements to the end

```python
# append()example

colors = ['White', 'red']
print(f"before appending: {colors}")

colors.append("blue")
print(f"after appending: {colors}")

colors.append("green")
print(f"after appending: {colors}")

nums = [3, 4, 5]
nums.append(6)

print(nums)
nums.append([7,8])
print(nums)
```

Extend
. Adding all elements from iterative to the end

```python
# extend examples
lis1 = [1,2,3]
lis2 = [4,5,6]
print(f"lis1:{lis1}")
print(f"lis2:{lis2}")
lis1.extend(lis2)
print(f"after extended:{lis2}")
lets = ['a','b']
lets.append('cd')
print(lets)
lets.extend('cd')
print(lets)
```

Pop()

```python
nums = [1,2,3,45,6]
print(nums)

#pop the element 
last_one = nums.pop()
print(f"popped :{last_one}")
print(f"After popping : {nums}")

#pop the specific index
second_one = nums.pop(1)
print(f"popped :{second_one}")
#print(f"After popping : {nums}")




```

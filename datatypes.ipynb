# Sequence Types

```python
#string (str)
Name = "python"
multiline = """ This is the class of python"""
print(f"{Name},type:{type (Name)}")
```

```python
#List (list) ordered and mutable
colors = ['blue','red','green']
mixed = [23,'shashi',True,[9,8],{"Name":"shashi"},("python"),89.0]
print(f"list : {colors},Type: {type(colors)}")
print(mixed)
```

```python
#Tuple (tuple) ordered,immutable
coordinates = (10,20)
one_num = (34,"hey")
print(f"tuple :{coordinates},Type:(type coordinates)")
print(one_num)
```

```python
#Range (range) -- sequence of numbers
number = range(5)
print(f"range : {list(number)}")
for i in range (10):
    print(i)
```

```python
number = range(5)
customized_num = range(2,10,3)
print(f"range:{list(number)}")
print(f"customized_num : {list( customized_num)}")
```

##Lists



-order - items maintain their position
-mutable - can be changed after creation
-Allow Duplicates some kind of value come many times
-Mixed Types - can be having multiple datatypes 

```python
list_methods = [i for i in dir(list) if not i.startswith('_')]
print(f"Total list methods:{len(list_methods)}\n")
print("All list methods")
for i, method in enumerate(list_methods, 1):
            print(f"{i:2}. {method}")
```

Creating list

```python
#Empty list
Empty_list = []
print(Empty_list)
```

```python
#list with values
nums = [4,6,8,9,3]
print(nums)
```

```python
#list with different types
mixed = [1,'hari', 3.89,True,None]
print(mixed)
range = list(range(5))
print(range)
dups = [1,2,2,3,4,1]
print(dups)
```

Accessing Elements in the list

```python
colors = ['red','blue','green','white']
print(colors)
print(f"colors[0]:{colors[0]}")
print(f"colors[3]:{colors[2]}")
print(f"colors[-1]:{colors[::-1]}")
```

Append

-- Adding elements to the end

```python
# append()example

colors = ['White', 'red']
print(f"before appending: {colors}")

colors.append("blue")
print(f"after appending: {colors}")

colors.append("green")
print(f"after appending: {colors}")

nums = [3, 4, 5]
nums.append(6)

print(nums)
nums.append([7,8])
print(nums)
```

Extend
. Adding all elements from iterative to the end

```python
# extend examples
lis1 = [1,2,3]
lis2 = [4,5,6]
print(f"lis1:{lis1}")
print(f"lis2:{lis2}")
lis1.extend(lis2)
print(f"after extended:{lis2}")
lets = ['a','b']
lets.append('cd')
print(lets)
lets.extend('cd')
print(lets)
```

insert()
.inserting elements at the specific index

```python
colors = ['green','red','white']
print(colors)
#inserting blue at 1 position
colors.insert(1,'blue')
print(f"after inserted:{colors}")
#inserting purple at o position
colors.insert(0,'purple')
print(f"after inserted: {colors}")
```

## Remove()
- removes the first occurance of the specific element

```python
colors = ['red','blue','green','yellow','white']
print(colors)
colors.remove('green')
print("After remove of the green{}".format(colors))
colors.remove('blue')
print("After remove of the blue{}".format(colors))

# mixed list you can create[] all the datatypes
#loop through then find if it type(i) == list remove('blue')
# add if there is a string the isalnum digit check
```

Pop()
.Remove and return element at index

```python
nums = [1,2,3,45,6]
print(nums)

#pop the element 
last_one = nums.pop()
print(f"popped :{last_one}")
print(f"After popping : {nums}")

#pop the specific index
second_one = nums.pop(1)
print(f"popped :{second_one}")
#print(f"After popping : {nums}")




```

Clear()
.Remove all elements for the list

```python
nums = [1,2,3,4,5,6]
print(nums)
nums.clear()
print(nums)
```

Index
.Returns the index of first occurance of an element

```python
colors = ['red','blue','white','green','yellow']
print(colors)

#Fines the index of an element 
bl_index = colors.index('blue')
print(f"index of blue : {bl_index}")
```

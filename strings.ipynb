Case Conversion Methods

```python
text = "hello python"

# capitalize() firstchar is upper the rest is lower
print ("capitalize(): {text.capitalize()}")
#upper() - All chars are uppercase
print(f"upper():{text.upper()}")
#lower() - All chars are lowercase
print(f"lower():{text.lower()}")
#title() - First char of each word is upper
print (f" title (): {text.title()}")
#swapcase() - swapping upper to lower and vice versa
print (f" swapcase(): {text.swapcase()}")
```

Alignment And Padding Methods

```python
text = "AgenticAI"
#center() center align with padding
print (f"center(20,'*') : {text.center(20,'*')}")
# ljust() left align with padding
print(f" , just(20,'_') : {text.ljust(20,'_')}")
# rjust() left align with padding
print(f", just (20.'_') : {text.ljust(20,'_')}")
```

Searching And Replacing Methods


```python
text = "hey , I am learning GenAI and AgenticAI, I love GenAI"
#find()
print(f"find (GenAI): {text.find('I')}")
#Index() like find the i[tendex if not the throws an error
print (f"Index (love) :{ text.index('GenAI')} ")
#count() how many times occurance
print(f"count (G):{text.index ('GenAI')}")
#replace
print(f"replace (GenAI , python): {text. replace ('GenAI', 'python')}")
```

Splitting and Joining Methods

```python
text = "hey ,iam learning GenAI and AgenticAI, I love GenAI"
#split split this string into a list
print (f"split:{text.split()}")
# rsplit() splilt from the right
print(f"rsplit(','): {text.rsplit(',',1)}")

#join() - join iterable with str separator
words = ('hey','Iam',"learning","GenAI")
",".join(words)
```

Trimming Methods

```python
text = "  iam learning Agentic AI"
#strip - remove learning and trailing spaces
print(f"strip():{text.strip()}")
#rstrip -- Remove leading characters
print(f"rstrip(): {text.rstrip()}")
#sstrip --remove trailing characters
print(f"sstrip(): {text.lstrip()}")
#remove prefix()remove prefix
url = "https://hari.com"
print(f"{url.removeprefix('https://')}")

filename = "hari.jpeg"
print(f"{url.removesuffix('.jpeg')}")
```
